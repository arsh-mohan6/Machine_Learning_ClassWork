```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
import numpy as np

def logistic_cost(X,y,w,b):
    m = X.shape[0]
    cost = 0.0
    for i in range(m):
        z_i = np.dot(X[i],w)+b
        f_wb_i = sigmoid(z_i)
        f_wb_i = np.clip(f_wb_i, 1e-15, 1 - 1e-15)


        cost+=y[i]*np.log(f_wb_i) + (1-y[i])*np.log(1-f_wb_i)
    cost = -cost/m
    return cost

def main():
    X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
    y_train = np.array([0, 0, 0, 1, 1, 1])

    w_tmp = np.array([1,1])
    b_tmp = -3
    print(logistic_cost(X_train, y_train, w_tmp, b_tmp).round(4))
main()
    
```
